```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('Bristol_Centre.csv')
df2 = pd.read_csv('Bristol_Temple_Way.csv')
df3 = pd.read_csv('Bristol_Old_Market.csv')
df4 = pd.read_csv('Bristol_St_Pauls.csv')

display(df2)
```

```python
files = ['Bristol_Centre.csv', 'Bristol_Temple_Way.csv', 'Bristol_Old_Market.csv', 'Bristol_St_Pauls.csv']

df = pd.DataFrame()
for file in files:
    # Read the CSV file into a dataframe
    temp = pd.read_csv(file)
    # Rename the columns if needed
    temp = temp.rename(columns={'Station Name': 'Station name'})
    # Concatenate the dataframe with the existing dataframe
    df = pd.concat([df, temp], axis=0, ignore_index=True, sort=False)
df = df.fillna('No data')
```

```python
df = df.sort_values(by=['Date', 'Time'])


df.to_csv('merged_file.csv', index=False)
```

```python
# display the data frame
display(df)
```

```python
df = pd.read_csv('merged_file.csv')
df.loc[df['Station name'] == 'No data', 'Station name'] = np.nan
df = df.dropna(subset=['Station name'])
df.to_csv('new_file.csv', index=False)
display(df)
```
